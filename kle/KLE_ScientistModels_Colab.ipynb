# KLE Scientist Models — Colab Runner

12 种科学方法 + 生成式记忆架构，从 GitHub 自动拉取运行。

**方法列表：** Information Theory, Markov, Graph Network, Bayesian, Poisson, SVD, UCB Bandit, Fourier, KNN, RNN, Genetic Algorithm, Ensemble Fusion

**模式：**
- `compare` — 12 方法单注对比（~10s）
- `predict` — 生成式记忆预测 N 组（~5min）
- `extreme` — 极限搜索大池化（~8min）
- `walkforward` — 带记忆的 walk-forward 验证（~6min）

```python
# ── Step 1: Clone repo & install deps ──
import os, subprocess

REPO = "https://github.com/jursmatsko/lotto-image-predictor-research.git"
DIR  = "lotto-image-predictor-research/kle"

if not os.path.exists(DIR):
    subprocess.run(["git", "clone", REPO], check=True)
else:
    subprocess.run(["git", "-C", DIR, "pull"], check=True)

os.chdir(DIR)
subprocess.run(["pip", "install", "-q", "pandas", "numpy", "scikit-learn"], check=True)
print("✅ Ready! Working dir:", os.getcwd())
```

## Step 2: 设定参数

修改下面的变量来控制运行模式和目标期号。

```python
# ── Parameters ──
TARGET_ISSUE = "2026040"   # 目标期号（数据中已有的可回看，未来的则盲测）
N_SETS       = 20          # 生成组数
POOL_SIZE    = 50000       # extreme 模式候选池大小
WF_EVAL      = 20          # walk-forward 评估期数
WF_WARMUP    = 15          # walk-forward 记忆预热期数
```

## Step 3: Compare — 12 方法单注对比（~10 秒）

```python
!python -m scripts.scientist_models compare --issue {TARGET_ISSUE}
```

## Step 4: Predict — 生成式记忆预测 N 组（~5 分钟）

```python
!python -m scripts.scientist_models predict --issue {TARGET_ISSUE} --sets {N_SETS}
```

## Step 5: Extreme — 极限搜索大池化（~8 分钟）

从数万组候选中找到命中最高的组合。

```python
!python -m scripts.scientist_models extreme --issue {TARGET_ISSUE} --sets {N_SETS} --pool-size {POOL_SIZE}
```

## Step 6: Walk-Forward — 带记忆验证（~6 分钟）

每期更新记忆，下一期用更新后的记忆预测。

```python
!python -m scripts.scientist_models walkforward --issue {TARGET_ISSUE} --eval {WF_EVAL} --warmup {WF_WARMUP} --sets {N_SETS}
```
